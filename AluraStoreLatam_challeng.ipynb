### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
def calcular_facturacion_total(tienda, nombre):
  total= tienda['Precio'].sum()
  print(f"üõí Ingreso total de {nombre}: ${total:,.2f}")
  return total
```

```python
facturacion_t1 = calcular_facturacion_total(tienda, 'Tienda 1')
facturacion_t2 = calcular_facturacion_total(tienda2, 'Tienda 2')
facturacion_t3 = calcular_facturacion_total(tienda3, 'Tienda 3')
facturacion_t4 = calcular_facturacion_total(tienda4, 'Tienda 4')
```

# 2. Ventas por categor√≠a

```python
def categorias_populares(tienda, nombre):
    resumen = tienda['Categor√≠a del Producto'].value_counts()
    print(f"\nüè∑Ô∏è Categor√≠as m√°s populares en {nombre}:")
    print(resumen)
    return resumen
```

```python
cat_t1 = categorias_populares(tienda, 'Tienda 1')
cat_t2 = categorias_populares(tienda2, 'Tienda 2')
cat_t3 = categorias_populares(tienda3, 'Tienda 3')
cat_t4 = categorias_populares(tienda4, 'Tienda 4')
```

# 3. Calificaci√≥n promedio de la tienda


```python
def promedio_calificacion(tienda, nombre):
    promedio = tienda['Calificaci√≥n'].mean()
    print(f"‚≠ê Calificaci√≥n promedio en {nombre}: {promedio:.2f}")
    return promedio

prom_t1 = promedio_calificacion(tienda, 'Tienda 1')
prom_t2 = promedio_calificacion(tienda2, 'Tienda 2')
prom_t3 = promedio_calificacion(tienda3, 'Tienda 3')
prom_t4 = promedio_calificacion(tienda4, 'Tienda 4')
```

# 4. Productos m√°s y menos vendidos

```python
def productos_mas_y_menos_vendidos(tienda, nombre):
    conteo = tienda['Producto'].value_counts()
    mas_vendido = conteo.idxmax()
    menos_vendido = conteo.idxmin()
    print(f"\nüìà En {nombre}:")
    print(f"üîù Producto m√°s vendido: {mas_vendido} ({conteo.max()} ventas)")
    print(f"üîª Producto menos vendido: {menos_vendido} ({conteo.min()} ventas)")
    return conteo
```

```python
ventas_t1 = productos_mas_y_menos_vendidos(tienda, 'Tienda 1')
ventas_t2 = productos_mas_y_menos_vendidos(tienda2, 'Tienda 2')
ventas_t3 = productos_mas_y_menos_vendidos(tienda3, 'Tienda 3')
ventas_t4 = productos_mas_y_menos_vendidos(tienda4, 'Tienda 4')
```

# 5. Env√≠o promedio por tienda

```python
def costo_envio_promedio(tienda, nombre):
    promedio = tienda['Costo de env√≠o'].mean()
    print(f"üì¶ Costo de env√≠o promedio en {nombre}: ${promedio:,.2f}")
    return promedio
```

```python
envio_t1 = costo_envio_promedio(tienda, 'Tienda 1')
envio_t2 = costo_envio_promedio(tienda2, 'Tienda 2')
envio_t3 = costo_envio_promedio(tienda3, 'Tienda 3')
envio_t4 = costo_envio_promedio(tienda4, 'Tienda 4')
```

# 6.  1. Gr√°fico de barras ‚Äì Facturaci√≥n total por tienda


```python
import matplotlib.pyplot as plt

facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()

facturacion = [facturacion_t1, facturacion_t2, facturacion_t3, facturacion_t4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8, 5))
plt.bar(tiendas, facturacion, color='skyblue')
plt.title('Facturaci√≥n total ($) por tienda')
plt.ylabel('Pesos colombianos')
plt.xlabel('Tiendas')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 6. Gr√°fico de dispersi√≥n: Calificaci√≥n vs. Costo de env√≠o


```python
calificaciones = [prom_t1, prom_t2, prom_t3, prom_t4]

plt.figure(figsize=(8, 5))
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', color='green')
plt.title(' Calificaci√≥n promedio por tienda')
plt.ylabel('Calificaci√≥n (1 a 5)')
plt.ylim(0, 5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
